### 1.3.4 리스트(list)
+ 인덱스를 기준으로 순차적으로 데이터 저장한다.

### 1.4.2 클래스(class)
+ 독자적인 자료형을 만들어낼 수 있다.
+ 해당 클래스만의 전용함수, 속성을 정의할 수 있다.

class 의 구조
```
class 클래스 이름:
    def __init__(self, 인수, ...): #  생성자
        ...
    def 메서드이름1(self, 인수, ...): # 메서드 1
        ...
    def 메서드이름2(self, dlstn, ...): # 메서드 2
        ...
```



### 생성자 ###
\__init\__ : 클래스의 인스턴스가 만들어질 때 한번만 불리며, 클래스를 초기화하는 방법을 정의한다.

파이썬에서는 메서드의 첫 번째 인수로 자신을 나타내는 self를 명시적으로 쓰는 특징이 있다.

+ Man이라는 __클래스__를 정의  
+ m이라는 __인스턴스(객체)__를 생성
+ name 이라는 인수를 받아 self.name을 초기화

````python
class Man:
    def __init__(self, name):
        self.name = name
        print("Initialized!")

    def hello(self):
        print("Hello "+self.name+"!")

    def goodbye(self):
        print("Good-bye "+self.name+"!")

m = Man("David")
m.hello()
m.goodbye()
````

### 1.5.3 넘파이의 산술 연산
+ 배열들의 원소 수를 맞춰서 산술 연산을 할 수 있다.

````python
x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
print(x + y)
print(x - y)
print(x * y)
print(x / y)
````

### 1.5.4 넘파이의 N차원 배열
+ 1차원, 2차원, 3차원 이상의 배열을 각각 백터, 행렬, 텐서 라고 부른다.(포함관계)
+ 차원의 크기가 같은 행렬끼리의 산술 연산은 대응하는 원소별로 계산된다.

shape : 배열의 차원의 크기를 알려주는 함수
dtype : 배열에 담긴 원소의 자료형을 알려주는 함수

````python
# 2차원 배열의 연산
A = np.array([[1, 2], [3, 4]])
print(A)
print(A.shape)
print(A.dtype)
````

````python
B = np.array([[3, 0], [0, 6]])
print(A + B)
print(A * B)
````

### 1.5.6원소 접근

````python
X = np.array([[51, 55], [14, 19], [0, 4]])
print(X)
print(X[0]) # 0행
print(X[0][1]) # (0,1) 위치의 원소
````

````python
# for문으로 원소에 접근
for row in X:
    print(row)
````

````python
X = X.flatten()                 # X를 1차원 배열로 평탄화
print(X)

print(X[np.array([0, 2, 4])])   # 원하는 인덱스 원소만 추출
````

````python
print(X>15)
print(X[X>15])      # 마스크 기능
````

# 1.6 matplotlib
+ 딥러닝에서는 데이터 시각화가 매우 중요하다.
+ 파이썬을 통해 그래프를 그려주는 라이브러리

````python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1 간격으로 생성
y1 = np.sin(x)
y2 = np.cos(x)

# 그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")    # sin, cos 함수 그리기, cos는 점선으로
plt.xlabel("x")
plt.ylabel("y")                                 # x, y축 이름
plt.title('sin & cos')                          # 제목
plt.legend()                                    # 범주 표시
plt.show()                                      # 그래프 보여주기
````
